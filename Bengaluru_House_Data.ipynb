# 1. Filling Null Values

```python
#replace NaN value to zero
df2=df.fillna(value=0)
```

```python
df2
```

```python
df2.isnull().sum()
```

```python
#replace NaN value to 5
df3=df.fillna(value=5)
```

```python
df3
```

```python
# Filling Null Values with a previous value

df4=df.fillna(method='pad')
```

```python
df4
```

```python
df.isnull().sum()
```

```python
# Filling Null values with next row value
df5=df.fillna(method='bfill')
df5
```

```python
# Filling Null values with next column value
df6=df.fillna(method='pad',axis=1)
df6
```

```python
# Filling Null values with next column value
df7=df.fillna(method='bfill',axis=1)
df7
```

```python
df
```

```python
# Filling Different Values in Null in diffrerent column

df8=df.fillna({'society':'abcd','balcony':'dfge'})
```

```python
df8
```

```python
# Filling Null Value with the mean of column

df9=df.fillna(value=df['balcony'].mean())
```

```python
df9
```
