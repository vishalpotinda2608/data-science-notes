```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
nf1_data=pd.read_csv('NFL Play by Play 2009-2016 (v3).csv')
```

```python
np.random.seed(0)
```

```python
nf1_data.head()
```

```python
# get the number of missing data points per column
total_missing_value_count=nf1_data.isnull().sum()
total_missing_value_count[0:10]
```

```python
#  how many total missing values do we have?
total_cells=np.product(nf1_data.shape)
total_missing=total_missing_value_count.sum()
```

```python
# Percent of data that is missing
percent_missing=(total_missing/total_cells)*100
print(percent_missing)
```

#### 24.8584% data is missing data

# Drop missing Values

```python
nf1_data.dropna()
```

```python
# remove all columns with at least one missing value
columns_with_na_dropped = nf1_data.dropna(axis=1)
columns_with_na_dropped.head()
```

```python
nf1_data.isnull().sum()
```

```python
# just how much data did we lose?
print("Columns in original dataset: %d \n" % nf1_data.shape[1])
print("Columns with na's dropped: %d" % columns_with_na_dropped.shape[1])
```

# Filling in missing values automatically

```python
#get a small subset of NFL dataset

subset_nfl_data=nf1_data.loc[:,"EPA":"Season"].head()
subset_nfl_data
```

```python
# Replace all NA'a with 0
subset_nfl_data.fillna(0)
```

```python
nf1_data.head()
```

```python
# replace all NA's the value that comes directly after it in the same column, 
# then replace all the remaining na's with 0
subset_nfl_data.fillna(method='bfill',axis=0).fillna(0)
```
